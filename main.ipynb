```python
# UNIVARIATE COX REGRESSION
results = {}
for i in genes.columns.tolist():
    if i != "Days" and i != "OS":
        cox.fit(genes[[i, "Days", "OS"]], 'Days', 'OS')
        results[i] = dict(cox.summary.iloc[0])
```

```python
import math
# IDENTIFY LOG HR
for i in results.keys():
    results[i]['log(HR)'] = math.log(results[i]['exp(coef)'], 10)
    results[i]['log_lower_95'] = math.log(results[i]['exp(coef) lower 95%'], 10)
    results[i]['log_upper_95'] = math.log(results[i]['exp(coef) upper 95%'], 10)
```

```python
results['PHF7']['p']
```

```python
# PLOT NORMAL DISTRIBUTIONS FOR EACH GENE AND ASSOCIATED Z-SCORE THRESHOLD
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the mean and standard deviation
mean = 0
std_dev = 1

# Generate x values
x = np.linspace(-4.5, 4.5, 1000)

# Generate y values using the PDF of the normal distribution
y = stats.norm.pdf(x, mean, std_dev)

for i in results.keys():
    z_score_threshold = abs(results[i]['z'])
    
    # Create the plot
    plt.figure(figsize = (6, 6))
    plt.plot(x, y, label='Normal Distribution (mean=0, std=1)')

    plt.axvline(x=z_score_threshold, color='r', linestyle='--', label=f'z-score = {z_score_threshold}')
    plt.axvline(x=-z_score_threshold, color='r', linestyle='--', label=f'z-score = {-z_score_threshold}')

    # Fill the area beyond the threshold
    plt.fill_between(x, y, where=(x >= z_score_threshold), color='red', alpha=0.3)
    plt.fill_between(x, y, where=(x <= -z_score_threshold), color='red', alpha=0.3)
    # Add title and labels

    plt.title(f'Normal Distribution (μ=0, σ=1) for {i} | p-value: {round(results[i]["p"], 5)}')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    # Show the legend
    plt.savefig(f"results/univariate/normal_models/{i}.png", dpi = 2000, bbox_inches = "tight")
```

```python
from lifelines import *
for j in sig_genes:
    if j == "RILPL2":
        median = genes[j].median(0)
        r = logrank_test(genes[genes[j] > median]['Days'], genes[genes[j] <= median]['Days'], genes[genes[j] > median]['OS'], genes[genes[j] <= median]['OS'])
        logrank_results[j] = [r.test_statistic, r.p_value, r.degrees_of_freedom]
        kmf = KaplanMeierFitter()
        kmf.fit(genes[genes[j] > median]['Days'], genes[genes[j] > median]['OS'])
        kmf2 = KaplanMeierFitter()
        kmf2.fit(genes[genes[j] <= median]['Days'], genes[genes[j] <= median]['OS'])
        plt.figure(figsize = (6, 6))
        kmf.plot_survival_function(label = f"High {j} Expression | n = {len(genes[genes[j] > median])}")
        kmf2.plot_survival_function(label = f"Low {j} Expression | n = {len(genes[genes[j] <= median])}") 
        plt.xlabel("Time from diagnosis (Days)")
        plt.ylabel("Proportion Surviving")
        plt.title(f"Survival Risk Stratification via {j} Expression")
        plt.savefig(f"results/risk_stratification/{j}.png", dpi = 2000)
```

```python
import numpy as np
q1, q2, q3 = np.percentile(gene_sig_df['risk_score'], [25, 50, 75])

group1 = gene_sig_df[gene_sig_df['risk_score'] <= q1]
group2 = gene_sig_df[(gene_sig_df['risk_score'] > q1) & (gene_sig_df['risk_score'] <= q2)]
group3 = gene_sig_df[(gene_sig_df['risk_score'] > q2) & (gene_sig_df['risk_score'] <= q3)]
group4 = gene_sig_df[gene_sig_df['risk_score'] > q3]

# Perform log-rank tests between all pairs of groups and store p-values
logrank_results = {}
logrank_results[f"G1_G2"] = logrank_test(group1['Days'], group2['Days'], group1['OS'], group2['OS']).p_value
logrank_results[f"G1_G3"] = logrank_test(group1['Days'], group3['Days'], group1['OS'], group3['OS']).p_value
logrank_results[f"G1_G4"] = logrank_test(group1['Days'], group4['Days'], group1['OS'], group4['OS']).p_value
logrank_results[f"G2_G3"] = logrank_test(group2['Days'], group3['Days'], group2['OS'], group3['OS']).p_value
logrank_results[f"G2_G4"] = logrank_test(group2['Days'], group4['Days'], group2['OS'], group4['OS']).p_value
logrank_results[f"G3_G4"] = logrank_test(group3['Days'], group4['Days'], group3['OS'], group4['OS']).p_value

# Fit Kaplan-Meier curves for each group
kmf1 = KaplanMeierFitter()
kmf1.fit(group1['Days'], group1['OS'], label=f"Q1 ({len(group1)})")

kmf2 = KaplanMeierFitter()
kmf2.fit(group2['Days'], group2['OS'], label=f"Q2 ({len(group2)})")

kmf3 = KaplanMeierFitter()
kmf3.fit(group3['Days'], group3['OS'], label=f"Q3 ({len(group3)})")

kmf4 = KaplanMeierFitter()
kmf4.fit(group4['Days'], group4['OS'], label=f"Q4 ({len(group4)})")
```
